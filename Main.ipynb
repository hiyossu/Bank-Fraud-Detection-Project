```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

data_dir = Path.cwd() / "data"

df = pd.read_csv('./data/bank_fraud.csv')
```

```python
df['transaction_timestamp'] = df['transaction_date'].astype(str) + ' ' + df['transaction_time'].astype(str)
df['transaction_timestamp'] = pd.to_datetime(df['transaction_timestamp'])

df['transaction_timestamp']
```

```python
fraud_counts = df['fraud_type'].value_counts()
fraud_counts = fraud_counts / 9000
print(fraud_counts)
```

```python
fraud_counts.plot.barh(stacked=True, figsize=(8,5))
plt.ylabel('types of fraud')
plt.title('fraud summary')
plt.show()
```
